```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X , y , coef = make_regression(
    n_samples=100,
    n_features=1,
    bias=3,
    noise=5,
    coef=True,
    random_state=10,
)
```

```python
print(X.shape , y.shape)
```

```python
M = np.linspace(-50 , 50 , 100)
C = np.linspace(-50 , 50 ,100)

print(M.shape , C.shape)

M, C = np.meshgrid(M,C)
print(M.shape , C.shape)

# we need to reshape it in 2D
M,C = M.reshape(-1,1) , C.reshape(-1 ,1)                    ############   1 step reshape it to have (-1,1)
print(M.shape , C.shape)
```

```python
def loss(X , y , M , C):
    y_hat = M*X + C
    residual_sq =( y - y_hat)**2
    J = residual_sq.sum(1)               ############### column wise addition
    return J
```

```python
J = loss(X, y , M , C)
print(J.shape)
```

```python
plt.figure(figsize = (7,7))
ax = plt.axes(projection='3d')
ax.scatter(M , C , J , c = J , cmap = plt.cm.coolwarm)
ax.set_xlabel("M")
ax.set_ylabel("C")
ax.set_zlabel("Z")
plt.show()
```
