```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
import numpy as np
```

```python
## create data

x1 , x2 = make_regression(n_samples = 200 ,n_features=1,bias = 30 ,noise = 15 ,random_state=43)
print(x1.shape , x2.shape)
```

```python
X = np.concatenate([x1,x2.reshape(-1,1)],axis=1)
print(X.shape)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

## principal components`

```python
## Centering data around mean
mean = X.mean(axis = 0)
print(mean)
X_centered = X - mean
print(X_centered.mean(axis=0))
```

```python
plt.scatter(X_centered[:,0],X_centered[:,1])
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
## eigen vectors
eigenvalues , eigenvectors = np.linalg.eig(X.T.dot(X))
print(eigenvalues)
print(eigenvectors)
```
