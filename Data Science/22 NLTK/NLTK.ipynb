## how to remove stop words from words?

```python
stop_words = stopwords.words('english')
```

```python
def remove_stopwords(text , sw):
    useful_words = []
    for word in text:
        if word.lower() in sw:
            continue
        else:
            useful_words.append(word)
    return useful_words
```

```python
usefulword = remove_stopwords(words , stop_words)
```

```python
len(usefulword)
```

## 3. Lemmetization and stemming

- it helps to get into root word
- ex forget , forgot , forgotten 
- root word - forget

- stemming and lemmetization both do the same work but in differnt way

```python
from nltk.stem import PorterStemmer , SnowballStemmer
```

```python
ps = PorterStemmer()
```

```python
ps.stem('establishment')
```

```python
ps.stem('establishing')
```

```python
def steming(words , stemmer):
    stemmed = []
    for words in words:
        stemmed.append(stemmer.stem(words))
    return stemmed
```

```python
stemmed_words = steming(usefulword , ps)
```
